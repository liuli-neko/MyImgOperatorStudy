```python
import matplotlib.pyplot as plt
import numpy as np
import cmath
```

```python
# exp(-2j*pi*(x/n + y/m))
# e^(ix)=(cos x+isin x)
def f(x,y,n,m):
    dx = -2j*np.pi*(x/n + y/m)
    return np.exp(dx)
```

```python
data_x = []
data_y = []
n,m = 100,20
for x in range(n):
    for y in range(m):
        cval = f(x,y,n,m)
        data_x.append(cval.real)
        data_y.append(cval.imag)
```

```python
plt.figure(figsize=(10,10),dpi=100)
plt.scatter(data_x, data_y)
plt.show()
```
